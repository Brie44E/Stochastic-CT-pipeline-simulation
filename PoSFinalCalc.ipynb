```python
# Create smaller DataFrames for each name with non-zero values and specific columns
individual_companies = {}
for column in pivot_company_df.columns:
    if column in ['DrugBank_Approval', 'Nature_Approvals'] or (pivot_company_df[column] != 0).any():
        non_zero_indexes = pivot_company_df.index[pivot_company_df[column] != 0]
        smaller_df = pivot_company_df.loc[non_zero_indexes, ['DrugBank_Approval', 'Nature_Approvals', column]]
        individual_companies[column] = smaller_df
```

```python
def PoSCalculator(TARGETDF):
    # Compute the total 
    approved_drugs = []
    approved_nature_durgs = []
    just_approved = []

    for index, row in TARGETDF.iterrows():
        if "Approved" in row["DrugBank_Approval"] and row["Nature_Approvals"] == None: #Generic_approval_name
            approved_drugs.append(index)
        elif row["Nature_Approvals"] != None: #Generic_approval_name
            approved_nature_durgs.append(index)
        if "Approved" == row["DrugBank_Approval"]:
            just_approved.append(index)
    return approved_drugs,approved_nature_durgs,just_approved
```

```python
companyList = []
numberOfID = []
numberOfPreApproved = []
numberOfApproved = []
PoS1 = []
PoS2 = []
just = []

for company,data in individual_companies.items():
    print(company)
    approved_drugs,approved_nature_durgs,just_approved = PoSCalculator(data)
    print(f"{company} conducted clinical trials on {len(data)} compounds found in DrugBank. Of those {len(approved_drugs)} were approved outside of the observation period")
    print(f"In total {len(approved_nature_durgs)} compounds were approved by the FDA (acc. to nature publications)")
    print(f"The PoS is {len(approved_nature_durgs)/(len(data)-len(approved_drugs))*100} or {len(approved_nature_durgs)/(len(data)-len(just_approved))*100} if only focused on the soley 'Approved' compounds")
    print()
    companyList.append(company)
    numberOfID.append(len(data))
    numberOfPreApproved.append(len(approved_drugs))
    numberOfApproved.append(len(approved_nature_durgs))
    PoS1.append(len(approved_nature_durgs)/(len(data)-len(approved_drugs))*100)
    PoS2.append(len(approved_nature_durgs)/(len(data)-len(just_approved))*100)
    just.append(len(just_approved))
```

```python
# Saving the data in a table
companyPoS = pd.DataFrame({'Companies': companyList, 'Total_IDs': numberOfID, 'Previously_Approved': numberOfPreApproved, 'Approved':numberOfApproved, 'PoS minus all approval':PoS1,'Number of just "Approved" assets': just, 'PoS minus "approval"': PoS2})
```

```python
# Printing out lists of DB IDs per company
def IDsPerCompany (company, dataset = boolean_df):
    Ids = []
    for index, row in dataset.iterrows():
        if row[company] != False:
            Ids.append(index)
    return Ids
```

```python
# Printing out Approved and Unapproved DB IDs

def approvalStatusPerID (company, dataset = boolean_df):
    natureApproved = []
    onlyDBapproved = []
    notApproved = []
    ids = IDsPerCompany(company)
    for id in ids: 
        if "Approval" in boolean_df.loc[id]["DrugBank_Approval"] and boolean_df.loc[id]["Nature_Approvaled_Generic"] == None:
            onlyDBapproved.append(id)
        elif boolean_df.loc[id]["Nature_Approvaled_Generic"] != None:
            natureApproved.append(id)
        else:
            notApproved.append(id)
    return natureApproved,onlyDBapproved,notApproved
```
