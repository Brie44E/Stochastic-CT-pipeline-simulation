```python
import requests
from bs4 import BeautifulSoup

url = "https://www.fda.gov/about-fda/histories-fda-regulated-products/summary-nda-approvals-receipts-1938-present"

# Make a GET request to the website
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the correct XPath for the table
    table_xpath = '//*[@id="main-content"]//div[@class="table-wrapper"]/table'

    # Extract and print the table content
    table = soup.select_one(table_xpath)
    
    if table:
        table_content = table.get_text(separator='\n')
        print(table_content)
    else:
        print("Table not found on the webpage.")

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://www.fda.gov/about-fda/histories-fda-regulated-products/summary-nda-approvals-receipts-1938-present"

# Set up Selenium with Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(options=chrome_options)

# Open the website
driver.get(url)

# Wait for the table to be present
table_xpath = '//*[@id="main-content"]/div/table'
table_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, table_xpath)))

# Extract the text content of the table
table_content_text = table_element.text

# Create a DataFrame from the table text
df = pd.read_html(table_content_text)[0]

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```
