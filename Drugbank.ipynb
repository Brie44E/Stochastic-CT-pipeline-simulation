```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

# Set up the Chrome driver  # Provide the path to your ChromeDriver executable
driver = webdriver.Chrome()
external_IDs = pd.Series()
titles = pd.Series()

for ID in df["DrugBank ID"]:
    # URL of the website
    url = f"https://go.drugbank.com/drugs/{ID}"

    # Open the URL
    driver.get(url)

    try:
        # Find the elements by class name
        elements = driver.find_elements(By.CLASS_NAME, "list-unstyled.inline-list")   

        # Extract text content of the element
        if element:
            # Join text content of each element into a single string
            text_content = ' '.join([element.text for element in elements])
            # Create a pandas Series from the joined text
            #series_data = pd.Series(text_content)
            #print(series_data)
            external_IDs.loc[len(external_IDs)] = text_content
            try:
                # Find the element using XPath
                element = driver.find_element(By.XPATH, '//*[@id="associated-therapies"]')
                title_content = element.text.strip()
                series_data = pd.Series(title_content)
                #print(series_data)
                titles.loc[len(titles)] = title_content
            except:
                titles.loc[len(titles)] = None
                
        else:
            print("Element not found.")
    except: 
        print(f"Couldn't find external IDs {ID}")
        external_IDs.loc[len(external_IDs)] = None
    # Quit the driver
driver.quit()
external_ID_check = pd.concat([titles, external_IDs], axis=1)
```

```python
%%time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas

# Set up the Chrome driver
driver = webdriver.Chrome()
#external_IDs = pd.DataFrame({'External_IDs'})
external_IDs = pd.Series()
strangeCases = []
df["External_IDs "] = ""
df["Brand_Names"] = ""
df["DB_Approval"] = ""

for index, row in df.iterrows():
    # URL of the website
    ID = row["DrugBank ID"]
    url = f"https://go.drugbank.com/drugs/{row["DrugBank ID"]}"

    # Open the URL
    driver.get(url)
    # getting the External IDs
    try:
        # Check if the element with ID 'external-ids' is present
        external_ids_element = driver.find_element(By.ID, 'external-ids')

        if external_ids_element:
            # If the element is present, find the next <dd> element and print its text
            next_dd_element = external_ids_element.find_element(By.XPATH, "./following-sibling::dd")
            #print(next_dd_element.text)
            value = next_dd_element.text
            df.loc[index, 'External_IDs'] = value
            # print(value)
        else:
            print(f"Element with ID 'external-ids' found but Error at {ID}")
            df.loc[index, 'External_IDs'] = None
            strangeCases.append(ID)
    except:
        #print(f"Element with ID 'external-ids' not found at {ID}")
        df.loc[index, 'External_IDs'] = None
    # getting the Brand Names 
    try:
        # Check if the element with ID 'Brand Names' is present
        external_ids_element = driver.find_element(By.ID, 'brand-names')

        if external_ids_element:
            # If the element is present, find the next <dd> element and print its text
            next_dd_element = external_ids_element.find_element(By.XPATH, "./following-sibling::dd")
            #print(next_dd_element.text)
            value = next_dd_element.text
            df.loc[index, 'Brand_Names'] = value
            # print(value)
        else:
            print(f"Element with ID {ID} has no 'Brand Names'")
            df.loc[index, 'Brand_Names'] = None
            #strangeCases.append(ID)
    except:
        #print(f"Element with ID 'external-ids' not found at {ID}")
        print(f"Element with ID {ID} has no 'Brand Names'")
        df.loc[index, 'Brand_Names'] = None
    # getting the Approvals
    try:
        # Check if the element with ID 'groups' is present
        external_ids_element = driver.find_element(By.ID, 'groups')

        if external_ids_element:
            # If the element is present, find the next <dd> element and print its text
            next_dd_element = external_ids_element.find_element(By.XPATH, "./following-sibling::dd")
            #print(next_dd_element.text)
            value = next_dd_element.text
            df.loc[index, 'DB_Approval'] = value
            # print(value)
        else:
            print(f"Element with ID 'external-ids' found but Error at {ID}")
            df.loc[index, 'DB_Approval'] = None
            strangeCases.append(ID)
    except:
        #print(f"Element with ID 'groups' not found at {ID}")
        df.loc[index, 'DB_Approval'] = None


# Quit the driver
driver.quit()
```
