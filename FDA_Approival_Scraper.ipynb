```python
# Get the FDA@Drugs table
def tableScraper(letter):
    url = f"https://www.accessdata.fda.gov/scripts/cder/daf/index.cfm?event=browseByLetter.page&productLetter={letter}"

    # Set up the Chrome WebDriver (update the path to the chromedriver executable)
    driver = webdriver.Chrome()

    # Navigate to the URL
    driver.get(url)

    try:
        # Find the table using XPATH
        table_xpath = '//*[@id="main-content"]/div/table'
        table_element = driver.find_element(By.XPATH, table_xpath)
        
        # Extract hrefs from the table
        hrefs = [a.get_attribute('href') for a in table_element.find_elements(By.TAG_NAME, 'a')]

        # Extract the HTML content of the table
        table_html = table_element.get_attribute('outerHTML')

        # Read the HTML content into a DataFrame
        df = pd.read_html(table_html)[0]

        #resulting_string = " ".join(df[0].tolist())

        # Print or use the DataFrame as needed
        return df, hrefs
    except:
        print (letter)
```

```python

def extractNDAandBLA(resulting_string):
    # Define a regex pattern to match "NDA #xxxxxx" or "ANDA #xxxxxx"
    pattern = re.compile(r'(NDA|ANDA|BLA) #(\d+)')

    # Extract matching substrings
    matching_substrings = pattern.findall(resulting_string)

    # Print the matching substrings
    return matching_substrings, len(matching_substrings)
```
